```python
expenses = []

def add_expense(amount, category):
    expenses.append({"amount": amount, "category": category})

add_expense(225.50, "Groceries")
add_expense(150.00, "Transportation")
add_expense(215.75, "Entertainment")
add_expense(5.25, "Coffee")

# Step 2: Summarize by category
def spending_by_category():
    result = {}
    for e in expenses:
        cat = e["category"]
        amt = e["amount"]
        result[cat] = result.get(cat, 0) + amt
    return result

# Step 3: Plot the chart
import matplotlib.pyplot as plt

totals = spending_by_category()
categories = list(totals.keys())
amounts = list(totals.values())

plt.figure(figsize=(6, 6))
plt.pie(amounts, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title("Spending by Category")
plt.axis('equal')
plt.show()

explode = [0.1 if cat == "Groceries" else 0 for cat in categories]
plt.pie(
    amounts,                      # positional arg 1
    labels=categories,            # keyword
    explode=explode,              # keyword
    autopct='%1.1f%%',            # keyword
    startangle=90                 # keyword
)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

# Step 1: Expense data
expenses = [
    {"amount": 225.50, "category": "Groceries"},
    {"amount": 150.00, "category": "Transportation"},
    {"amount": 215.75, "category": "Entertainment"},
    {"amount": 5.25, "category": "Coffee"},
    {"amount": 60.00, "category": "Groceries"},
]

# Step 2: Create a DataFrame
df = pd.DataFrame(expenses)

# Step 3: Group by category
totals = df.groupby("category")["amount"].sum().sort_values(ascending=False)

# Step 4: Display tables
print("Detailed Expenses:")
display(df)

print("\nTotal Spending by Category:")
display(totals)

# Step 5: Plot charts side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Bar chart
axs[0].bar(totals.index, totals.values)
axs[0].set_title("Bar Chart")
axs[0].set_ylabel("Amount ($)")

# Pie chart
axs[1].pie(totals.values, labels=totals.index, autopct='%1.1f%%', startangle=90)
axs[1].set_title("Pie Chart")
axs[1].axis('equal')

plt.suptitle("Spending Dashboard", fontsize=16)
plt.tight_layout()
plt.show()
```
