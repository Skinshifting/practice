```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Expense data with dates
expenses = []

def add_expense(amount, category, date):
    expenses.append({"amount": amount, "category": category, "date": date})

add_expense(225.50, "Groceries", "2024-05-01")
add_expense(150.00, "Transportation", "2024-05-02")
add_expense(215.75, "Entertainment", "2024-05-03")
add_expense(5.25, "Coffee", "2024-05-03")
add_expense(60.00, "Groceries", "2024-05-04")

# Step 2: Create DataFrame
df = pd.DataFrame(expenses)
df["date"] = pd.to_datetime(df["date"])

# Step 3: Group by date and sum amounts
daily_totals = df.groupby("date")["amount"].sum()

# Step 4: Plot the line chart
plt.figure(figsize=(8, 5))
plt.plot(daily_totals.index, daily_totals.values, marker='o')
plt.title("Spending Over Time")
plt.xlabel("Date")
plt.ylabel("Total Spent ($)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

# Step 1: Expense data
expenses = [
    {"amount": 225.50, "category": "Groceries"},
    {"amount": 150.00, "category": "Transportation"},
    {"amount": 215.75, "category": "Entertainment"},
    {"amount": 5.25, "category": "Coffee"},
    {"amount": 60.00, "category": "Groceries"},
]

# Step 2: Create a DataFrame
df = pd.DataFrame(expenses)

# Step 3: Group by category
totals = df.groupby("category")["amount"].sum().sort_values(ascending=False)

# Step 4: Display tables
print("Detailed Expenses:")
display(df)

print("\nTotal Spending by Category:")
display(totals)

# Step 5: Plot charts side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Bar chart
axs[0].bar(totals.index, totals.values)
axs[0].set_title("Bar Chart")
axs[0].set_ylabel("Amount ($)")

# Pie chart
axs[1].pie(totals.values, labels=totals.index, autopct='%1.1f%%', startangle=90)
axs[1].set_title("Pie Chart")
axs[1].axis('equal')

plt.suptitle("Spending Dashboard", fontsize=16)
plt.tight_layout()
plt.show()
```
